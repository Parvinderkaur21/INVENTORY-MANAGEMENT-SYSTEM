***READ DATA FROM JSON***

```python
import json

fd = open("record.json","r")
r = fd.read()
fd.close()

record = json.loads(r)
```

***ADD NEW ITEMS TO THE INVENTORY SYSTEM***

```python
print("Welcome to the warehouse of MODERN STORE")
print("Please enter the details of the new product")

prod_id     = input("Enter product id: ")
product     = input("Enter name of the product: ")
price       = int(input("Enter price if the product: "))
quantity    = int(input("Enter no. of products: "))
measurement = input("What is the measurement of product - kilo/piece/litre: ")
supplier    = input("Enter supplier of the product: ")
expirydate  = input("Enter the expiry date of the product:")


          
record[prod_id] = {"product" : product , "price": price , "quantity": quantity, "measurement": measurement , "expiry date": expirydate , "supplier": supplier}

js = json.dumps(record)

fd = open("record (3).json","w")
fd.write(js)
fd.close()
```
